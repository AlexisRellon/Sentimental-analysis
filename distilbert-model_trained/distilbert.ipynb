```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
import emoji  # Add emoji library for better preprocessing
import os

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

# Install additional packages if needed
!pip install emoji --quiet

df = pd.read_csv('spotify_reviews.csv')
df

# Enhanced Data Cleaning with better emoji handling

def preprocess_text(text):
    if pd.isna(text):  # Handle NaN values
        return ""
    # Convert to lowercase
    text = text.lower()
    # Handle emojis - convert to text representation
    text = emoji.demojize(text)
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove mentions and hashtags
    text = re.sub(r'\@\w+|\#\w+', '', text)
    # Remove special characters but preserve emoji text representations
    text = re.sub(r'[^\w\s\:\_\-]', '', text)
    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    # Tokenize
    tokens = word_tokenize(text)
    # Remove stopwords but keep negation words which are important for sentiment
    stop_words = set(stopwords.words('english')) - {'no', 'not', 'nor', 'neither', 'never', 'none'}
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return ' '.join(filtered_tokens)

# Only process content once and store it in processed_content
df['processed_content'] = df['content'].apply(preprocess_text)

# Filter out very short reviews after preprocessing
df = df[df['processed_content'].str.len() > 5]

# More nuanced sentiment mapping based on star rating
def stars_to_sentiment(stars):
    if stars <= 2:
        return 'negative'
    elif stars == 3:
        return 'neutral'
    else:
        return 'positive'

df['label'] = df['score'].apply(stars_to_sentiment)

print(f"Data shape after preprocessing: {df.shape}")
print(f"Label distribution: {df['label'].value_counts()}")

# Display examples from each class for verification
for label in ['negative', 'neutral', 'positive']:
    print(f"\nExample {label} review:")
    print(df[df['label'] == label].iloc[0]['processed_content'])
    print(f"Original: {df[df['label'] == label].iloc[0]['content']}")

# Rest of imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import get_linear_schedule_with_warmup
import random
from torch.optim import AdamW
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm

def set_seed(seed_value=42):
    random.seed(seed_value)
    np.random.seed(seed_value)
    torch.manual_seed(seed_value)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed_value)

set_seed()

# Improved data augmentation for better balance
def augment_text(text, strength=0.3):
    """Augment text with controlled strength parameter"""
    if pd.isna(text) or len(text) < 5:
        return text

    words = text.split()
    if len(words) < 4:  # Skip very short texts
        return text

    # 1. Random word deletion (with controlled strength)
    if random.random() < 0.5:
        words = [w for w in words if random.random() > strength]

    # 2. Random word swap (30% chance)
    if len(words) > 3 and random.random() < 0.3:
        swap_count = max(1, int(len(words) * strength * 0.5))
        for _ in range(swap_count):
            idx1, idx2 = random.sample(range(len(words)), 2)
            words[idx1], words[idx2] = words[idx2], words[idx1]

    return ' '.join(words) if words else text

# Get class distribution
class_counts = df['label'].value_counts()
print(f"Original class distribution: {class_counts}")

# Improved balancing strategy
max_class_size = class_counts.max()
balanced_dfs = []

# Process each class
for label, count in class_counts.items():
    class_df = df[df['label'] == label]

    if count < max_class_size:
        # Calculate how many augmented samples we need
        samples_needed = max_class_size - count
        augment_factor = samples_needed // count + 1

        # Create augmented samples
        aug_dfs = []
        for strength in np.linspace(0.2, 0.4, augment_factor):
            aug_df = class_df.copy()
            aug_df['processed_content'] = aug_df['processed_content'].apply(
                lambda x: augment_text(x, strength=strength)
            )
            aug_dfs.append(aug_df)

        # Combine original and augmented data
        combined_df = pd.concat([class_df] + aug_dfs, ignore_index=True)
        # Sample to get exact number needed
        balanced_df = combined_df.sample(max_class_size, replace=False if len(combined_df) >= max_class_size else True)
    else:
        # For majority class, just use original samples
        balanced_df = class_df.sample(max_class_size, replace=False)

    balanced_dfs.append(balanced_df)

# Combine all balanced classes
balanced_df = pd.concat(balanced_dfs, ignore_index=True)
print(f"Balanced data shape: {balanced_df.shape}")
print(f"Balanced label distribution: {balanced_df['label'].value_counts()}")

#####################################
# SAVE THE BALANCED DATA TO CSV
#####################################

# Save the processed [augmented data] to a csv
balanced_df.to_csv('spotify_reviews_balanced.csv', index=False)

#####################################
# PREPARE MODEL
#####################################

# Train-test split
train_df, val_df = train_test_split(balanced_df, test_size=0.2, stratify=balanced_df['label'], random_state=42)

# Visualize class balance before proceeding
plt.figure(figsize=(10, 6))
sns.countplot(x='label', data=balanced_df)
plt.title('Class Distribution After Balancing')
plt.ylabel('Count')
plt.xlabel('Sentiment Label')
plt.xticks(rotation=0)
plt.show()

# Model
class SentimentDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length=128):  # Increased max_length for better context
        self.data = dataframe
        self.tokenizer = tokenizer
        self.max_length = max_length
        # Create a mapping from sentiment labels to numerical indices
        self.label_mapping = {'negative': 0, 'neutral': 1, 'positive': 2}

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        # Use processed_content instead of content
        text = self.data.iloc[index]['processed_content']
        label = self.data.iloc[index]['label']

        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )

        # Convert label to numerical index using the mapping
        label_index = self.label_mapping[label]

        # Convert to expected format
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label_index, dtype=torch.long)  # Use label_index
        }

# Choose a model - keeping distilbert for efficiency
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
train_dataset = SentimentDataset(train_df, tokenizer)
val_dataset = SentimentDataset(val_df, tokenizer)

# Increased batch size for faster training, with batch size 32 for training as well
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Improved evaluation function that handles errors correctly
def evaluate(model, dataloader):
    model.eval()
    predictions = []
    actual_labels = []
    eval_loss = 0
    num_batches = 0

    with torch.no_grad():
        for batch in dataloader:
            # Move batch to device
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            # Forward pass with labels for loss calculation
            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)

            # Accumulate loss
            eval_loss += outputs.loss.item()
            num_batches += 1

            # Get predictions
            preds = torch.argmax(outputs.logits, dim=1).cpu().numpy()
            actual = labels.cpu().numpy()

            predictions.extend(preds)
            actual_labels.extend(actual)

    # Calculate metrics
    avg_loss = eval_loss / num_batches
    accuracy = accuracy_score(actual_labels, predictions)
    f1 = f1_score(actual_labels, predictions, average='weighted')
    class_names = ["negative", "neutral", "positive"]
    report = classification_report(actual_labels, predictions, target_names=class_names, digits=3)
    cm = confusion_matrix(actual_labels, predictions)

    return avg_loss, accuracy, f1, report, cm, predictions, actual_labels

# Proper class weights calculation
class_weights = {}
label_to_idx = {'negative': 0, 'neutral': 1, 'positive': 2}

# Count samples per class in original dataset
class_counts = df['label'].value_counts().to_dict()
total_samples = sum(class_counts.values())

# Calculate weights inversely proportional to class frequencies
for label, count in class_counts.items():
    weight = total_samples / (len(class_counts) * count)
    class_weights[label_to_idx[label]] = weight

# Convert to tensor
class_weights_tensor = torch.tensor([class_weights[i] for i in range(3)], dtype=torch.float32).to(device)

# Better optimizer configuration
optimizer = AdamW(model.parameters(), lr=2e-5, weight_decay=0.01, eps=1e-8)

# Training parameters
num_epochs = 10
total_steps = len(train_loader) * num_epochs
warmup_steps = int(0.1 * total_steps)
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=warmup_steps, num_training_steps=total_steps)

# For mixed precision training
from torch.amp import autocast, GradScaler
scaler = GradScaler()  # Remove the 'cuda' parameter as it's incorrect
accumulation_steps = 2  # Gradient accumulation for larger effective batch size

# Early stopping parameters
early_stopping_patience = 2
best_val_f1 = 0
patience_counter = 0

# For tracking metrics
train_losses = []
val_losses = []
train_accuracies = []
val_accuracies = []
val_f1_scores = []

#####################################
# START TRAINING MODEL
#####################################

# Start training loop with progress bar
print("Starting training...")
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    # Use tqdm for progress tracking
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}")

    for step, batch in enumerate(progress_bar):
        # Zero gradients only at the beginning of accumulation steps
        if step % accumulation_steps == 0:
            optimizer.zero_grad()

        # Move batch to device
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        # Forward pass with mixed precision - use correct autocast initialization
        with autocast(device_type='cuda' if torch.cuda.is_available() else 'cpu'):
            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                labels=labels
            )

            # Calculate weighted loss using the class weights
            loss = outputs.loss

            # Accuracy calculation
            logits = outputs.logits
            preds = torch.argmax(logits, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

        # Scale loss for gradient accumulation
        loss = loss / accumulation_steps

        # Backward pass with mixed precision
        scaler.scale(loss).backward()

        # Update weights at accumulation step boundary
        if (step + 1) % accumulation_steps == 0 or (step + 1 == len(train_loader)):
            # Gradient clipping
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

            scaler.step(optimizer)
            scaler.update()
            scheduler.step()

        # Update progress bar
        total_loss += loss.item() * accumulation_steps
        current_loss = total_loss / (step + 1)
        current_acc = correct / total if total > 0 else 0

        # Get current validation metrics (only for display in the first batch of each epoch)
        current_val_loss = val_losses[-1] if val_losses else 0
        current_val_acc = val_accuracies[-1] if val_accuracies else 0

        # Update progress bar with all metrics
        progress_bar.set_postfix({
            'train_loss': f"{current_loss:.4f}",
            'train_acc': f"{current_acc:.4f}",
            'val_loss': f"{current_val_loss:.4f}",
            'val_acc': f"{current_val_acc:.4f}"
        })

    # Calculate training metrics for this epoch
    epoch_train_loss = total_loss / len(train_loader)
    epoch_train_acc = correct / total if total > 0 else 0
    train_losses.append(epoch_train_loss)
    train_accuracies.append(epoch_train_acc)

    # Evaluate at the end of each epoch
    print(f"\nEvaluating after epoch {epoch+1}...")
    val_loss, val_acc, val_f1, report, cm, _, _ = evaluate(model, val_loader)
    val_losses.append(val_loss)
    val_accuracies.append(val_acc)
    val_f1_scores.append(val_f1)

    print(f"Epoch {epoch+1}/{num_epochs}")
    print(f"Training Loss: {epoch_train_loss:.4f}, Accuracy: {epoch_train_acc:.4f}")
    print(f"Validation Loss: {val_loss:.4f}, Accuracy: {val_acc:.4f}, F1: {val_f1:.4f}")
    print(f"Classification Report:\n{report}")

    # Early stopping check
    if val_f1 > best_val_f1:
        best_val_f1 = val_f1
        patience_counter = 0
        # Save best model
        torch.save(model.state_dict(), "best_spotify_sentiment_model.pt")
        print("Saved best model!")
    else:
        patience_counter += 1
        print(f"No improvement for {patience_counter} epochs")

    if patience_counter >= early_stopping_patience:
        print(f"Early stopping triggered after {epoch+1} epochs")
        break

#####################################
# SAVE THE MODEL (SERIALIZATION)
#####################################

import joblib
# Save trained model with joblib
joblib.dump(model, "sentiment_model.pkl")
joblib.dump(tokenizer, "tokenizer.pkl")

#####################################
# LOAD MODEL AND SHOW TRAINING RESULT
#####################################

# Load best model for final evaluation
model.load_state_dict(torch.load("best_spotify_sentiment_model.pt"))

# Plot training metrics
plt.figure(figsize=(12, 10))

# Plot training and validation loss
plt.subplot(2, 2, 1)
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title('Loss vs. Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot training and validation accuracy
plt.subplot(2, 2, 2)
plt.plot(train_accuracies, label='Training Accuracy')
plt.plot(val_accuracies, label='Validation Accuracy')
plt.title('Accuracy vs. Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Plot F1 score
plt.subplot(2, 2, 3)
plt.plot(val_f1_scores, label='Validation F1 Score', color='green')
plt.title('F1 Score vs. Epochs')
plt.xlabel('Epochs')
plt.ylabel('F1 Score')
plt.legend()

# Final confusion matrix heatmap
plt.subplot(2, 2, 4)
_, _, _, _, cm, _, _ = evaluate(model, val_loader)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Negative', 'Neutral', 'Positive'],
            yticklabels=['Negative', 'Neutral', 'Positive'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')

plt.tight_layout()
plt.savefig('training_metrics.png')
plt.show()

# Final evaluation
print("\nFinal Evaluation:")
val_loss, val_acc, val_f1, report, cm, predictions, actual_labels = evaluate(model, val_loader)
print(f"Final Loss: {val_loss:.4f}")
print(f"Final Accuracy: {val_acc:.4f}")
print(f"Final F1 Score: {val_f1:.4f}")
print(f"Final Classification Report:\n{report}")
print(f"Final Confusion Matrix:\n{cm}")

# Save model
output_dir = "./spotify_sentiment_model"
model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)
print(f"Model saved to {output_dir}")

#####################################
# TEST MODEL
#####################################

# Test Model with improved visualization
test_texts = [
    "This app is terrible, it keeps crashing and wasting my time.",
    "The application works as expected, nothing special to mention.",
    "I absolutely love this application! Best music experience ever!",
    "Average app, does what it says but has some bugs.",
    "Not worth the money, terrible customer service and full of glitches."
]

# Function to create color-coded sentiment display
def display_sentiment_prediction(text, sentiment, confidence):
    color = {
        'negative': 'red',
        'neutral': 'blue',
        'positive': 'green'
    }[sentiment]

    from IPython.display import HTML
    return HTML(f"""
    <div style="margin-bottom: 20px;">
        <p><strong>Text:</strong> {text}</p>
        <p><strong>Sentiment:</strong> <span style="color:{color};font-weight:bold;">{sentiment}</span></p>
        <p><strong>Confidence:</strong> <progress value="{confidence}" max="100"></progress> {confidence:.1f}%</p>
    </div>
    """)

# Preprocessing
model.eval()
sentiment_map = {0: "negative", 1: "neutral", 2: "positive"}

print("\nTesting model on sample texts:")
results = []

for text in test_texts:
    # Use the same preprocessing as during training
    processed_text = preprocess_text(text)
    inputs = tokenizer(
        processed_text,
        return_tensors="pt",
        truncation=True,
        padding="max_length",
        max_length=128
    ).to(device)

    with torch.no_grad():
        outputs = model(**inputs)

    probs = torch.nn.functional.softmax(outputs.logits, dim=1)
    prediction = torch.argmax(outputs.logits, dim=1).item()
    confidence = probs[0][prediction].item() * 100

    print(f"Text: {text}")
    print(f"Sentiment: {sentiment_map[prediction]} (Confidence: {confidence:.1f}%)\n")

    # Save results for visualization
    results.append({
        'text': text,
        'sentiment': sentiment_map[prediction],
        'confidence': confidence
    })

# Create a nice visualization of results
plt.figure(figsize=(10, 8))
sentiments = [r['sentiment'] for r in results]
confidences = [r['confidence'] for r in results]
colors = ['red' if s == 'negative' else 'blue' if s == 'neutral' else 'green' for s in sentiments]

bars = plt.barh(range(len(test_texts)), confidences, color=colors)
plt.yticks(range(len(test_texts)), [t[:30] + '...' if len(t) > 30 else t for t in test_texts])
plt.xlabel('Confidence (%)')
plt.title('Sentiment Analysis Results')

# Add a legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='red', label='Negative'),
    Patch(facecolor='blue', label='Neutral'),
    Patch(facecolor='green', label='Positive')
]
plt.legend(handles=legend_elements)

plt.tight_layout()
plt.savefig('sentiment_results.png')
plt.show()

print("Analysis complete!")

# Save results to CSV for further analysis
results_df = pd.DataFrame(results)
results_df.to_csv('sentiment_analysis_results.csv', index=False)
print("Results saved to sentiment_analysis_results.csv")
```
